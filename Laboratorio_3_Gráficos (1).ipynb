2. Construye una tabla de frecuencias absolutas y relativas (puntuales y acumuladas) para la variable *Edad*, considera 10 intervalos.

```python
import pandas as pd

df_1=pd.read_excel("/content/BDActividad_1_2_3L3.xlsx")
muestra_3=df_1.shape[0]
df_1['Intervalo_Edad'] = pd.cut(df_1['Edad'],bins=10,include_lowest=True, precision=0)
cuenta_edad=df_1.groupby("Intervalo_Edad", observed =True).size()
tabla_3=pd.DataFrame({
  "Frecuencia Absoluta": cuenta_edad,
  "Frecuencia Absoluta acumulada": cuenta_edad.cumsum(),
  "Frecuencia Relativa (%)": cuenta_edad/muestra_3 *100,
  "Frecuencia Relativa (%) Absoluta":(cuenta_edad/muestra_3 * 100).cumsum()
})
tabla_3
```

4. Construye un gráfico apropiado para la variable *Edad* y justifica su elección. Luego, interpreta la mayor y menor frecuencia, y el rango del  histograma.

```python
import matplotlib.pyplot as plt

categorias_edad = tabla_3.index
f_edad = tabla_3['Frecuencia Relativa (%)']
fig,aux = plt.subplots(figsize = (7,5))
histogram_edad = aux.bar(categorias_edad.astype(str), f_edad, color = 'green',edgecolor='black', width=1)
aux.set_title('Distribucion de trabajadores por edad')
aux.set_xlabel('Edad (años)')
aux.set_ylabel('cantidad de trabajadores')
aux.bar_label(histogram_edad,fmt="%.1f%%")
plt.xticks(rotation=45,ha = "right")
plt.show()
```
